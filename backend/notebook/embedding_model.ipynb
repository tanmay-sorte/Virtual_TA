```python
import os
from pymongo import MongoClient

RESUME_BASE_DIR = r"C:\Users\TSorte\OneDrive - Rockwell Automation, Inc\Desktop\Virtual_TA (TA-AI)\resumes" # Update this to your actual base path

def _list_all_jobs():
    client = MongoClient("mongodb://localhost:27017")  # Update if needed
    db = client["virtual_ta"]
    jobs_col = db["jobs"]

    # Fetch jobs that have a requirement_id
    jobs = list(jobs_col.find({"requirement_id": {"$exists": True, "$ne": None}}))

    # Filter jobs where folder exists
    jobs_with_folders = []
    for job in jobs:
        req_id = job.get("requirement_id")
        folder_path = os.path.join(RESUME_BASE_DIR, req_id)
        if os.path.isdir(folder_path):
            jobs_with_folders.append(job)

    return jobs_with_folders
```
