````python
def score_resume_fields(jd_json : dict, resume_json : dict):
    prompt = f"""
    Compare these two profiles:

        JD : {jd_json}
        Resume : {resume_json}

    Score (0-100) for:
        - Skill Match
        - Education Match
        - Experience Match

    Then calulate weighted overall score : (Skills: 0.7, Education: 0.1, Experience: 0.2).
    Return JSON with each field score and final score.

    While finding field wise score focus on complete resume rather than field from resume.
    Also use semantic matching rather than keywords. Also where skills are mentioned, try
    to find context or application of the skill, and give more advantage if application or 
    context is found.
    """

    response = client.chat.completions.create(
        model= os.getenv("DEPLOYMENT"),
        messages=[{"role" : "user", "content" : prompt}]
    )

    return response.choices[0].message.content
````

````python
def safe_cosine(a, b) -> float:
    a = np.asarray(a, dtype=float)
    b = np.asarray(b, dtype=float)
    denom = np.linalg.norm(a) * np.linalg.norm(b)
    if denom == 0:
        return 0.0
    return float(np.dot(a, b) / denom)

def strip_json(text: str) -> str:
    """
    Strip markdown fences and try to extract a JSON object.
    """
    if not text:
        return "{}"
    # Remove triple backticks blocks
    text = re.sub(r"^```(?:json)?\s*|\s*```$", "", text.strip(), flags=re.MULTILINE)
    # Try to find the first {...} JSON object
    m = re.search(r"\{.*\}", text, flags=re.DOTALL)
    return m.group(0) if m else text


def skills_to_text(skills):
    """
    Accepts dict{category: [skills]} | list[str] | str | None
    Returns a readable string for embedding.
    """
    if not skills:
        return ""
    if isinstance(skills, str):
        return skills
    if isinstance(skills, dict):
        parts = []
        for category, items in skills.items():
            if not items:
                continue
            line = f"{category}: " + "; ".join(str(x) for x in items)
            parts.append(line)
        return " | ".join(parts)
    if isinstance(skills, (list, tuple)):
        return ", ".join(str(x) for x in skills)
    return str(skills)

def education_to_text(education_list):
    """
    Expects list[dict] like {"degree": "...", "institution": "...", "year"/"dates": "..."}
    """
    if not education_list:
        return ""
    lines = []
    for edu in education_list:
        if isinstance(edu, dict):
            degree = edu.get("degree")
            inst = edu.get("institution") or edu.get("institute") or edu.get("university")
            year = edu.get("year") or edu.get("dates")
            segments = [seg for seg in [degree, inst, year] if seg]
            if segments:
                lines.append(" â€” ".join(segments))
        else:
            lines.append(str(edu))
    return " || ".join(lines)

def experience_to_text(experience_list):
    """
    Expects list[dict] like {"title": "...", "company": "...", "dates": "...", "responsibilities": list[str]}
    """
    if not experience_list:
        return ""
    lines = []
    for exp in experience_list:
        if isinstance(exp, dict):
            title = exp.get("title")
            company = exp.get("company")
            dates = exp.get("dates")
            header_parts = [p for p in [title, company, dates] if p]
            header = " | ".join(header_parts) if header_parts else ""
            resp = exp.get("responsibilities")
            if isinstance(resp, (list, tuple)):
                resp_text = "; ".join(str(r) for r in resp)
            elif resp is not None:
                resp_text = str(resp)
            else:
                resp_text = ""
            if header and resp_text:
                lines.append(f"{header} â€” {resp_text}")
            elif header:
                lines.append(header)
            elif resp_text:
                lines.append(resp_text)
        else:
            lines.append(str(exp))
    return " || ".join(lines)
````

````python
def compute_similarity(inputs):
    # Expect keys: jd_embeddings (with 'skills','education','experience'), resumes: list[{file, content}]
    jd_emb = inputs["jd_embedding"]
    jd_emb_skill = np.asarray(jd_emb.get("skills", []), dtype=float)
    jd_emb_edu   = np.asarray(jd_emb.get("education", []), dtype=float)
    jd_emb_exp   = np.asarray(jd_emb.get("experience", []), dtype=float)

    results = []
    for item in inputs.get("resumes", []):
        content = item.get("content", {})

        resume_skills = skills_to_text(content.get("Skills"))
        resume_edu    = education_to_text(content.get("Education"))
        resume_exp    = experience_to_text(content.get("Experience"))

        res_emb_skill = get_embeddings(resume_skills)
        res_emb_edu   = get_embeddings(resume_edu)
        res_emb_exp   = get_embeddings(resume_exp)

        sim_skill = safe_cosine(jd_emb_skill, res_emb_skill)
        sim_edu   = safe_cosine(jd_emb_edu, res_emb_edu)
        sim_exp   = safe_cosine(jd_emb_exp, res_emb_exp)

        # Convert to percentage (0-100), round 2 decimals
        sim_score_skill = round(sim_skill * 100, 2)
        sim_score_edu   = round(sim_edu * 100, 2)
        sim_score_exp   = round(sim_exp * 100, 2)

        # Optional overall score (weights can be tuned)
        overall = round(0.6 * sim_score_skill + 0.1 * sim_score_edu + 0.3 * sim_score_exp, 2)

        results.append({
            "file": item.get("file"),
            "content": content,
            "normalized_text": {
                "skills": resume_skills,
                "education": resume_edu,
                "experience": resume_exp,
            },
            "similarity": {
                "skills": sim_score_skill,
                "education": sim_score_edu,
                "experience": sim_score_exp,
                "overall": overall,
            },
        })

    return {
        "jd": inputs.get("jd"),
        "jd_embeddings": inputs.get("jd_embeddings"),
        "results": results,
        "job_id": inputs.get("job_id"),
    }
````

````python
import os
import re
from typing import Dict, Any, Tuple, List

# --- Add at top of your file ---
def _read_pdf_pymupdf(path: str):
    debug = {"engine": "pymupdf", "pages": 0, "chars": 0, "flags": []}
    try:
        import fitz  # PyMuPDF
    except Exception as e:
        debug["flags"].append(f"pymupdf_not_available: {e}")
        return "", debug

    text_parts = []
    with fitz.open(path) as doc:
        debug["pages"] = doc.page_count
        for page in doc:
            t = page.get_text("text") or ""
            text_parts.append(t)
    full_text = "\n".join(text_parts)
    debug["chars"] = len(full_text)
    return full_text, debug

def _read_pdf_pdfminer(path: str):
    debug = {"engine": "pdfminer", "pages": None, "chars": 0, "flags": []}
    try:
        from pdfminer.high_level import extract_text
    except Exception as e:
        debug["flags"].append(f"pdfminer_not_available: {e}")
        return "", debug
    try:
        text = extract_text(path) or ""
        debug["chars"] = len(text)
        return text, debug
    except Exception as e:
        debug["flags"].append(f"pdfminer_extract_error: {e}")
        return "", debug

def _read_pdf_robust(path: str):
    # Try PyMuPDF first
    text, d1 = _read_pdf_pymupdf(path)
    if d1.get("chars", 0) > 0:
        return text, {"pdf": d1}

    # Fallback: pdfminer
    text2, d2 = _read_pdf_pdfminer(path)
    combo_debug = {"pymupdf": d1, "pdfminer": d2}
    if len(text2) > 0:
        return text2, {"pdf": combo_debug}

    # Nothing worked; likely scanned or unreadable without OCR
    flags = []
    flags.extend(d1.get("flags", []))
    flags.extend(d2.get("flags", []))
    return "", {"pdf": combo_debug, "flags": list(set(flags))}

# --- Then in load_resume(), replace your PDF branch with: ---



# ---- File-type readers ----
def _read_txt(path: str) -> str:
    with open(path, "r", encoding="utf-8", errors="ignore") as f:
        return f.read()

def _read_pdf(path: str) -> Tuple[str, Dict[str, Any]]:
    """
    Extracts text using PyMuPDF. Returns (text, debug).
    If text is empty/near empty, it likely needs OCR (scanned PDF).
    """
    debug = {"pages": 0, "chars": 0, "flags": []}
    try:
        import fitz  # PyMuPDF
    except Exception as e:
        debug["flags"].append(f"pymupdf_not_available: {e}")
        return "", debug

    text_parts = []
    with fitz.open(path) as doc:
        debug["pages"] = doc.page_count
        for page in doc:
            t = page.get_text("text")
            # If text extraction fails, try a different extractor mode:
            if not t.strip():
                t = page.get_text("blocks") or ""
                # 'blocks' returns list/str depending on version; coerce to str:
                if isinstance(t, list):
                    t = "\n".join([" ".join(map(str, b)) for b in t])
            text_parts.append(t)
    full_text = "\n".join(text_parts)
    debug["chars"] = len(full_text)
    if debug["pages"] > 0 and debug["chars"] < 20:
        debug["flags"].append("likely_scanned_pdf_no_text")
    return full_text, debug
def _read_docx(path: str) -> Tuple[str, Dict[str, Any]]:
    debug = {"paras": 0, "chars": 0, "flags": []}
    try:
        import docx  # python-docx
    except Exception as e:
        debug["flags"].append(f"python-docx_not_available: {e}")
        return "", debug
    try:
        doc = docx.Document(path)
        paras = [p.text for p in doc.paragraphs]
        debug["paras"] = len(paras)
        text = "\n".join(paras)
        debug["chars"] = len(text)
        return text, debug
    except Exception as e:
        debug["flags"].append(f"docx_parse_error: {e}")
        return "", debug

# ---- Field extractors (heuristic) ----
EMAIL_RE = re.compile(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}")
PHONE_RE = re.compile(r"(?:\+\d{1,3}[- ]?)?(?:\(?\d{3,4}\)?[- ]?)?\d{3}[- ]?\d{3,4}")

# Very simple name heuristic: top few lines without digits/specials
def _extract_name(text: str) -> str:
    lines = [l.strip() for l in text.splitlines()]
    lines = [l for l in lines if l]
    candidates = []
    for l in lines[:8]:  # check first 8 non-empty lines
        if len(l) > 3 and not any(ch.isdigit() for ch in l):
            # Avoid lines that look like headings with symbols
            if sum(ch.isalpha() for ch in l) / max(1, len(l)) > 0.6:
                candidates.append(l)
    # Prefer the earliest reasonably short candidate
    candidates = sorted(candidates, key=lambda x: len(x))
    return candidates[0] if candidates else ""

def _extract_email(text: str) -> str:
    m = EMAIL_RE.search(text)
    return m.group(0) if m else ""

def _extract_phone(text: str) -> str:
    # Find a reasonable-length phone
    phones = PHONE_RE.findall(text)
    if not phones:
        return ""
    # Return the first with at least 10 digits total
    for ph in phones:
        digits = re.sub(r"\\D", "", ph)
        if len(digits) >= 10:
            return ph
    return phones[0]

def _extract_skills(text: str, skill_keywords: List[str]) -> List[str]:
    text_lower = text.lower()
    found = []
    for kw in skill_keywords:
        # simple contains; you can improve with tokenization
        if kw.lower() in text_lower:
            found.append(kw)
    return sorted(set(found))

def _extract_education(text: str) -> str:
    edu_keywords = [
        "b.e", "btech", "b.tech", "bachelor", "be ", "bs ", "m.e", "mtech", "m.tech",
        "ms ", "master", "bsc", "msc", "bca", "mca", "computer science", "engineering"
    ]
    t = text.lower()
    hits = [k for k in edu_keywords if k in t]
    if not hits:
        return ""
    # return a brief joined view
    return ", ".join(sorted(set(hits)))[:120]

def _extract_experience_years(text: str) -> float:
    """
    Very rough: looks for patterns like 'X years' or 'X+ years'.
    """
    years_re = re.compile(r"(\\d+(?:\\.\\d+)?)\\s*\\+?\\s*years", re.IGNORECASE)
    matches = [float(m.group(1)) for m in years_re.finditer(text)]
    return max(matches) if matches else 0.0

# ---- Main loader ----
def load_resume(path: str) -> Dict[str, Any]:
    """
    Returns a dict with fields:
      {
        'ok': bool,
        'reason': str,
        'file': str,
        'raw_text_len': int,
        'raw_head': str,  # first 300 chars
        'fields': {
            'name': str,
            'email': str,
            'phone': str,
            'education': str,
            'skills': [str],
            'experience_years': float
        },
        'debug': {...}
      }
    """
    if not path or not os.path.exists(path):
        return {"ok": False, "reason": f"file_not_found: {path}", "file": path}

    ext = os.path.splitext(path)[1].lower()
    debug = {"ext": ext, "flags": []}
    raw_text = ""

    try:
        if ext == ".txt":
            raw_text = _read_txt(path)
        elif ext == ".pdf":
            raw_text, pdf_debug = _read_pdf_robust(path)
            debug.update(pdf_debug)
        elif ext == ".docx":
            raw_text, docx_debug = _read_docx(path)
            debug.update({"docx": docx_debug})
        else:
            return {"ok": False, "reason": f"unsupported_extension: {ext}", "file": path}
    except Exception as e:
        return {"ok": False, "reason": f"read_error: {e}", "file": path, "debug": debug}

    raw_text = raw_text or ""
    raw_len = len(raw_text)

    if raw_len < 10:
        # Likely scanned PDF or empty document
        if ext == ".pdf":
            debug["flags"].append("empty_or_non_extractable_pdf_text")
            # Tip: Use OCR (pytesseract + tesseract OCR) to handle scanned PDFs.
        return {
            "ok": False,
            "reason": "no_extractable_text",
            "file": path,
            "raw_text_len": raw_len,
            "raw_head": raw_text[:300],
            "debug": debug
        }

    # --- Extract fields ---
    email = _extract_email(raw_text)
    phone = _extract_phone(raw_text)
    name = _extract_name(raw_text)

    # Customize skill dictionary per JD
    # Rockwell MES/Java flavor, trimmed for demo
    skill_dict = [
        "Java", "OOP", "Object-Oriented Design", "Design Patterns", "Data Structures",
        "Algorithms", "JUnit", "Maven", "Jenkins", "GitLab", "Eclipse",
        "SQL", "SQL Server", "Oracle", "Agile", "SAFe", "CI/CD",
        "JavaScript", "Vaadin", "Swing", "JIRA", "SVN", "ANT", "Docker",
        "Kubernetes", "TDD", "BDD", "Jenkins", "Microservices", "REST"
    ]
    skills = _extract_skills(raw_text, skill_dict)
    education = _extract_education(raw_text)
    exp_years = _extract_experience_years(raw_text)

    return {
        "ok": True,
        "reason": "parsed",
        "file": path,
        "raw_text_len": raw_len,
        "raw_head": raw_text[:300],
        "fields": {
            "name": name,
            "email": email,
            "phone": phone,
            "education": education,
            "skills": skills,
            "experience_years": exp_years
        },
        "debug": debug
    }
````

````python
import os
from typing import Dict, Any
from PyPDF2 import PdfReader
import docx

IGNORED_BASENAMES = {"jd_data", "jd_embedding"}

def load_resumes(inputs: Dict[str, Any]) -> Dict[str, Any]:
    resume_dir = inputs.get("resume_dir")
    if not os.path.isdir(resume_dir):
        # No resumes found; keep going so pipeline can still run
        return {**inputs, "resumes": []}

    resumes = []
    for filename in os.listdir(resume_dir):
        fullpath = os.path.join(resume_dir, filename)
        if not os.path.isfile(fullpath):
            continue

        # ignore hidden files
        if filename.startswith("."):
            continue

        base, ext = os.path.splitext(filename)
        # ignore jd_data / jd_embedding (any ext)
        if base.lower() in IGNORED_BASENAMES:
            continue

        ext = ext.lower()
        if ext not in [".txt", ".pdf", ".doc", ".docx"]:
            continue

        text = ""
        try:
            if ext == ".txt":
                with open(fullpath, "r", encoding="utf-8", errors="ignore") as f:
                    text = f.read()
            elif ext == ".pdf":
                reader = PdfReader(fullpath)
                text = "\n".join(page.extract_text() or "" for page in reader.pages)
            elif ext in [".doc", ".docx"]:
                doc = docx.Document(fullpath)
                text = "\n".join(p.text for p in doc.paragraphs)
        except Exception as e:
            print(f"Error reading {filename}: {e}")
            continue

        resumes.append({"file": filename, "text": text})

    print("load_resumes complete")
    return {**inputs, "resumes": resumes}
````

````python
def strip_to_json(text: str) -> str:
    """
    Remove Markdown fences and extract the first JSON object-ish payload.
    """
    if not text:
        return "{}"
    # Strip triple backticks if present
    text = text.strip()
    if text.startswith("```"):
        text = re.sub(r"^```(?:json)?\s*|\s*```$", "", text, flags=re.MULTILINE).strip()
    # Try to find a JSON object
    m = re.search(r"\{.*\}", text, flags=re.DOTALL)
    return m.group(0) if m else text
````

````python
def extract_resume_fields(inputs: ScoringState) -> ScoringState:
    """
    For each resume text, ask the LLM to extract:
      - Name (string)
      - Skills (object or list)
      - Education (list)
      - Experience (list)
    """
    extracted_resumes = []
    for resume in inputs["resumes"]:
        prompt = (
            "Extract the following fields from the resume and return a STRICT JSON object with keys:\n"
            "Name (string), Skills (object or list), Education (list), Experience (list).\n\n"
            f"Resume:\n{resume['text']}\n"
        )

        # If your Azure model supports JSON response format, uncomment below for more reliable JSON:
        # response = client.chat.completions.create(
        #     model=os.getenv("DEPLOYMENT"),
        #     response_format={"type": "json_object"},
        #     messages=[{"role": "user", "content": prompt}],
        #     temperature=0.2,
        # )

        response = client.chat.completions.create(
            model=os.getenv("DEPLOYMENT"),
            messages=[{"role": "user", "content": prompt}],
        )

        raw = response.choices[0].message.content or "{}"
        raw_json = strip_to_json(raw)
        try:
            resume_data = json.loads(raw_json)
        except Exception:
            # Fallback to avoid crashing pipeline
            resume_data = {"Name": None, "Skills": None, "Education": None, "Experience": None}

        extracted_resumes.append({"file": resume["file"], "resume": resume_data})

    print("extract resume complete")

    return {**inputs, "resumes": extracted_resumes}
````

````python
def extract_text_from_file(base_path: str, filename : str):
    extensions = [".txt", ".pdf", ".docx"]
    for ext in extensions:
        path = os.path.join(base_path, filename + ext)
        
        if not os.path.exists(path):
            continue

        if ext == ".txt":
            try:
                with open(path, "r") as f:
                    return f.read()
            except:
                continue

        if ext == ".pdf":
            reader = PdfReader(path)
            return "\n".join(page.extract_text() or "" for page in reader.pages)
        
        if ext == ".docx":
            doc = docx.Document(path)
            return "\n".join(p.text for p in doc.paragraphs)
````

````python
def load_resume(inputs : InsightsState) -> InsightsState:
    resume_dir = os.path.join(BASE_RESUME_PATH, inputs["job_id"])

    inputs["resume_ext"] = extract_text_from_file(resume_dir, inputs["applicant_id"])

    return inputs
````
